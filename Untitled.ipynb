```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Healthcaree.csv', encoding='unicode_escape')
```

```python
df.drop(['Follow_Up_Required'],axis=1,inplace=True)
```

```python
df.columns
```

```python
pd.isnull(df).sum()
```

```python
df[['Name', 'Age', 'Diagnosis', 'Treatment', 'Admission_Date', 'Discharge_Date', 'Treatment_Cost_USD']].describe(include='all')
```

```python
ax = sns.countplot(x='Treatment',data=df)
sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.dtypes
```

```python
Diagnosis_treat = df.groupby('Diagnosis', as_index=False)['Treatment'].count().sort_values(by='Treatment', ascending=False)

sns.set(rc={'figure.figsize': (20,5)})
sns.barplot(x='Diagnosis', y='Treatment', data=Diagnosis_treat)
```

```python
sns.countplot(x='Doctor', hue='Treatment', data=df)

```

```python
df.shape
```

```python
Diagnosis_treat = df.groupby('Diagnosis', as_index=False)['Treatment_Cost_USD'].count().sort_values(by='Treatment_Cost_USD', ascending=False)

sns.set(rc={'figure.figsize': (20,5)})
sns.barplot(x='Diagnosis', y='Treatment_Cost_USD', data=Diagnosis_treat)
```

```python
Diagnosis_treat = df.groupby('Doctor', as_index=False)['Patient_Satisfaction_Score'].count().sort_values(by='Patient_Satisfaction_Score', ascending=False)

sns.set(rc={'figure.figsize': (20,5)})
sns.barplot(x='Doctor', y='Patient_Satisfaction_Score', data=Diagnosis_treat)
```

```python
Diagnosis_treat = df.groupby('Insurance_Coverage_%', as_index=False)['Treatment_Cost_USD'].count().sort_values(by='Treatment_Cost_USD', ascending=False)

sns.set(rc={'figure.figsize': (25,5)})
sns.barplot(x='Insurance_Coverage_%', y='Treatment_Cost_USD', data=Diagnosis_treat)
```
